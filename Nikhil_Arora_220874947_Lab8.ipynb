```python
def print_student_info():
    print("Name: Nikhil Arora")
    print("Student ID: 220874947")
    print("Section A or B")
    print("Email: narora46@my.yorku.ca")

def task0():
    print_student_info()
    
def is_sorted(a_list):
    return all(a_list[i] <= a_list[i + 1] for i in range(len(a_list) - 1))

def task1():
    x1 = [1, 4, 5, 9, 0, 8, 10]
    x2 = [1, 2, 4, 5, 6, 7, 9]
    x3 = []
    
    print(is_sorted(x1))
    print(is_sorted(x2))
    print(is_sorted(x3))
    
def merge_dict(dict1, dict2):
    for key, value in dict2.items():
        if key not in dict1:
            dict1[key] = value

def task2():
    dict1 = {8:"Exercise", 9:"Breakfast", 12:"Lunch", 3:"Study", 6:"Netflix"}
    dict2 = {8:"Sleep", 10:"Lab", 12:"Class", 4:"Call Mom"}
    
    print("dict1")
    print(dict1)
    print("dict2")
    print(dict2)

    merge_dict(dict1, dict2)

    print("dict2 merged into dict1")
    print(dict1)

def invert_dict(a_dict):
    return {v: k for k, v in a_dict.items()}

def task3():
    a_dict = {1:'one', 2:'two', 3:'three', 4:'four', 5:'five'}
    
    print(a_dict)
    inverted_dict = invert_dict(a_dict)
    print(inverted_dict)

def list_to_dict(a_list):
    return {i: a_list[i] for i in range(len(a_list))}

def task4():
    my_list = [1, "hello", 9.99, ["EECS", "1015"], {1:"1", 2:"2"}]
    
    print(my_list)
    dict_from_list = list_to_dict(my_list)
    print(dict_from_list)

def str_list_to_num_list(a_list):
    for i in range(len(a_list)):
        a_list[i] = int(a_list[i])
    
def task5():
    x = ['1', '2', '3', '4', '5', '6', '7', '8', '9']
    
    print(x)
    str_list_to_num_list(x)
    print(x)
    
def merge_lists(list1, list2):
    if not is_sorted(list1):
        raise ValueError("Assertion Error: List 1 is not sorted!")
    if not is_sorted(list2):
        raise ValueError("Assertion Error: List 2 is not sorted!")

    result = []
    i = j = 0

    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            result.append(list1[i])
            i += 1
        else:
            result.append(list2[j])
            j += 1

    result.extend(list1[i:])
    result.extend(list2[j:])
    return result

def task6():   
    while True:
        list1 = input("Input 1st sorted list of numbers [x1 x2 ...]: ").split()
        list1 = [int(x) for x in list1]
        
        list2 = input("Input 2nd sorted list of numbers [y1 y2 ...]: ").split()
        list2 = [int(y) for y in list2]

        try:
            merged_list = merge_lists(list1, list2)
            print("Merged list:")
            print(merged_list)
            repeat = input("Try again [Y/N]? ").lower()
            if repeat != 'y':
                break
        except ValueError as e:
            print(e)
            

def main():
    print("\n---- Task 1: Check if list is sorted ----")
    task1()
    print("\n---- Task 2: Merge dictionaries ----")
    task2()
    print("\n---- Task 3: Invert dictionaries ----")
    task3()
    print("\n---- Task 4: List to dictionary ----")
    task4()
    print("\n---- Task 5: String list to num list ----")
    task5()
    print("\n---- Task 6: Merge list with assert ----")
    task6()

if __name__ == "__main__":
    main()
```
